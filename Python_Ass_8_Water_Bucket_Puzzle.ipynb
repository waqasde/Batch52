```python
# Water Bucket Puzzle by Hassan Ali Khan

def display_buckets(buckets):
    # Display the current water levels in each bucket
    print("\nTry to get 4L of water into one of these buckets:\n")
    for level in range(8, 0, -1):
        print(f"{level}|", end=" ")
        print("WWWWWW" if buckets[8] >= level else "      ", end=" ")
        print(f"       {level}|", end=" ")
        print("WWWWWW" if buckets[5] >= level else "      ", end=" ")
        print(f"       {level}|", end=" ")
        print("WWWWWW" if buckets[3] >= level else "      ", end=" ")
        print()

    print(" +------+         +------+         +------+")
    print("    8L               5L               3L")


def fill_bucket(buckets, bucket_size):
    # Fill the selected bucket
    buckets[bucket_size] = bucket_size
    print(f"\nFilled the {bucket_size}L bucket.")


def empty_bucket(buckets, bucket_size):
    # Empty the selected bucket
    buckets[bucket_size] = 0
    print(f"\nEmptied the {bucket_size}L bucket.")


def pour_bucket(buckets, from_bucket, to_bucket):
    # Pour water from one bucket to another
    total_water = buckets[from_bucket] + buckets[to_bucket]
    transfer_amount = min(total_water, to_bucket)
    buckets[to_bucket] = transfer_amount
    buckets[from_bucket] = total_water - transfer_amount
    print(f"\nPoured water from the {from_bucket}L bucket into the {to_bucket}L bucket.")


def water_bucket_puzzle():
    # Define the initial state of the buckets (all empty)
    buckets = {8: 0, 5: 0, 3: 0}

    while True:
        # Display the current state of the buckets
        display_buckets(buckets)

        # Check if the player has won the game
        if 4 in buckets.values():
            print("\nCongratulations! You've solved the puzzle.")
            break

        # Ask the player for the action
        print("\nYou can:")
        print(" (F)ill the bucket")
        print(" (E)mpty the bucket")
        print(" (P)our one bucket into another")
        print(" (Q)uit")
        action = input("> ").lower()

        if action == 'q':
            print("\nThanks for playing!")
            break
        elif action == 'f':
            bucket_size = int(input("Select a bucket 8, 5, 3: "))
            if bucket_size in buckets:
                fill_bucket(buckets, bucket_size)
            else:
                print("\nInvalid bucket size.")
        elif action == 'e':
            bucket_size = int(input("Select a bucket 8, 5, 3: "))
            if bucket_size in buckets:
                empty_bucket(buckets, bucket_size)
            else:
                print("\nInvalid bucket size.")
        elif action == 'p':
            from_bucket = int(input("Select the bucket to pour from (8, 5, 3): "))
            to_bucket = int(input("Select the bucket to pour into (8, 5, 3): "))
            if from_bucket in buckets and to_bucket in buckets and from_bucket != to_bucket:
                pour_bucket(buckets, from_bucket, to_bucket)
            else:
                print("\nInvalid bucket choice.")
        else:
            print("\nInvalid action. Please try again.")


# Run the water bucket puzzle
water_bucket_puzzle()
```
